```python
import numpy as np 
import pandas as pd 
import os 
import random 
import shutil 
```

```python
def append_string_to_filenames(folder_path, append_str):
    # Loop through each file in the folder
    for filename in os.listdir(folder_path):
        # Construct the old file path
        old_file_path = os.path.join(folder_path, filename)
        
        # Skip directories
        if os.path.isdir(old_file_path):
            continue
        
        # Construct the new file name
        name, ext = os.path.splitext(filename)
        new_filename = f"{name}{append_str}{ext}"
        
        # Construct the new file path
        new_file_path = os.path.join(folder_path, new_filename)
        
        # Rename the file
        os.rename(old_file_path, new_file_path)

# rename files 
append_string_to_filenames("../output/replicator_markers_1/segmentation","_rm_1") 
```

```python
def get_random_png_samples(folder_path, frac_samples):
    # List all files in the folder
    all_files = os.listdir(folder_path)
    
    # Filter out only .png and .jpg files
    image_files = [file for file in all_files if file.endswith(('.png', '.jpg'))]
    
    # Randomly sample the specified number of files
    num_samples = np.floor(frac_samples * len(image_files)).astype(int)
    sampled_files = random.sample(image_files, num_samples)
    
    return sampled_files

def move_files(file_list, src_folder, dest_folder):
    # Ensure the destination folder exists
    os.makedirs(dest_folder, exist_ok=True)
    
    # Move each file in the list to the destination folder
    for filename in file_list:
        src_path = os.path.join(src_folder, filename)
        dest_path = os.path.join(dest_folder, filename)
        shutil.move(src_path, dest_path)

# random sample of images in folder  
set = 1 
dir_images = f"../output/replicator_markers_{set}/images" 
dir_masks = f"../output/replicator_markers_{set}/segmentation" 
frac_samples = 0.0001


rand_images_list = get_random_png_samples(dir_images, frac_samples) 
rand_masks_list = rand_images_list.copy() 
for index, filename in enumerate(rand_masks_list): 
    rand_masks_list[index] = filename.replace(".jpg", ".png") 

# go through list and move the images/segmentations and yaml files  
# out_dir = "./data/train/images"
# move_files(rand_files_list, dir, out_dir) 
```

```python
def list_files_in_directory(directory_path):
    # List all files in the directory
    files = os.listdir(directory_path)
    return files

def split_files_randomly(file_list, fraction):
    # Calculate the number of files for the first list
    num_files_first_list = int(len(file_list) * fraction)
    
    # Randomly shuffle the file list
    random.shuffle(file_list)
    
    # Split the file list into two lists
    first_list = file_list[:num_files_first_list]
    second_list = file_list[num_files_first_list:]
    
    return first_list, second_list

def copy_files(file_list, src_folder, dest_folder):
    # Ensure the destination folder exists
    os.makedirs(dest_folder, exist_ok=True)
    
    # Copy each file in the list to the destination folder
    for filename in file_list:
        src_path = os.path.join(src_folder, filename)
        dest_path = os.path.join(dest_folder, filename)
        shutil.copy(src_path, dest_path)

set = 3 
dir_images = f"../output/replicator_markers_{set}/images" 
dir_masks = f"../output/replicator_markers_{set}/segmentation" 
out_dir_images_train = "./data/train/images" 
out_dir_masks_train = "./data/train/masks" 
out_dir_images_val = "./data/val/images" 
out_dir_masks_val = "./data/val/masks" 
frac_samples_train = 0.8

list_images = list_files_in_directory(dir_images) 
train_images_list, val_images_list = split_files_randomly(list_images, frac_samples_train) 

train_masks_list = train_images_list.copy() 
train_yaml_list = train_images_list.copy() 
for index, mask in enumerate(train_masks_list): 
    train_masks_list[index] = mask.replace(".jpg", ".png") 
for index, yaml in enumerate(train_yaml_list):
    train_yaml_list[index] = yaml.replace(".jpg", ".yaml")    

val_masks_list = val_images_list.copy() 
val_yaml_list = train_images_list.copy() 
for index, mask in enumerate(val_masks_list): 
    val_masks_list[index] = mask.replace(".jpg", ".png") 
for index, yaml in enumerate(val_yaml_list):
    val_yaml_list[index] = yaml.replace(".jpg", ".yaml")    

# print(train_images_list) 
# print(train_masks_list)
# print(train_yaml_list)

# move files to train and val directories 
copy_files(train_images_list, dir_images, out_dir_images_train)
copy_files(train_masks_list, dir_masks, out_dir_masks_train)
# move_files(train_yaml_list, dir_images, out_dir_images)

copy_files(val_images_list, dir_images, out_dir_images_val)
copy_files(val_masks_list, dir_masks, out_dir_masks_val)
# move_files(train_yaml_list, dir_images, out_dir_images)
```

```python
# read all images in a folder, resize the image, and save in a new folder

import os
from PIL import Image

# Define input and output directories
input_dir = './data/val_images'
output_dir = './data/val_images_resized_2'
new_size = (480, 270)  # Define the new size

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Loop through all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  # Check for image files
        # Open the image
        img_path = os.path.join(input_dir, filename)
        img = Image.open(img_path)
        
        # Resize the image
        img_resized = img.resize(new_size)
        
        # Save the resized image to the output directory
        output_path = os.path.join(output_dir, filename)
        img_resized.save(output_path)

print("All images have been resized and saved.")
```
